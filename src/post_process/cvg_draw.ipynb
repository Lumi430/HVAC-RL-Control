```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
cvg_hist_file = "~/Documents/HVAC-RL-Control/src/a3c-res-v0.1/converge_summary.csv"
```

```python
df = pd.read_csv(cvg_hist_file)
```

```python
df
```

```python
x = range(41)
fig = plt.figure(figsize = (16, 8))
gs1 = gridspec.GridSpec(2, 3)
gs1.update(wspace=0.22, hspace=0.15)
ftsz = 22

ax1 = plt.subplot(gs1[0])
ax1.set_xlim(0, 41)
ax1.plot(x, df['case1'], 'k', label='Case1 (act1)')
ax1.plot(x, df['case4'], 'k--', label='Case4 (act2)')
ax1.plot(x, df['case7'], 'k:', label='Case7 (act3)')
ax1.legend(fontsize = 14)

plt.ylabel('Reward', fontsize = ftsz)
plt.title('lambda=0.6, Lmt_ppd=0.15', fontsize = 18)

ax2 = plt.subplot(gs1[1])
ax2.set_xlim(0, 41)
ax2.plot(x, df['case2'], 'k', label='Case2 (act1)')
ax2.plot(x, df['case5'], 'k--',label='Case5 (act2')
ax2.plot(x, df['case8'], 'k:', label='Case8 (act3)')
ax2.legend(fontsize = 14)
plt.title('lambda=0.6, Lmt_ppd=0.30', fontsize = 18)

ax3 = plt.subplot(gs1[2])
ax3.set_xlim(0, 41)
ax3.plot(x, df['case3'], 'k', label='Case3 (act1)')
ax3.plot(x, df['case6'], 'k--',label='Case6 (act2)')
ax3.plot(x, df['case9'], 'k:', label='Case9 (act3)')
ax3.legend(fontsize = 14)
plt.title('lambda=0.6, Lmt_ppd=1.0', fontsize = 18)

ax4 = plt.subplot(gs1[3])
ax4.set_xlim(0, 41)
ax4.plot(x, df['case10'], 'k', label='Case10 (act1)')
ax4.plot(x, df['case13'], 'k--',label='Case13 (act2)')
ax4.plot(x, df['case16'], 'k:', label='Case16 (act3)')
ax4.legend(fontsize = 14)
plt.xlabel('Training Step', fontsize = ftsz)
plt.ylabel('Reward', fontsize = ftsz)
plt.xticks(range(0, 41, 10), ['0', '5M', '10M', '15M', '20M'], fontsize = 18)
plt.title('lambda=0.4, Lmt_ppd=0.15', fontsize = 18)

ax5 = plt.subplot(gs1[4])
ax5.set_xlim(0, 41)
ax5.plot(x, df['case11'], 'k', label='Case11 (act1)')
ax5.plot(x, df['case14'], 'k--',label='Case14 (act2)')
ax5.plot(x, df['case17'], 'k:', label='Case17 (act3)')
ax5.legend(fontsize = 14)
plt.xlabel('Training Step', fontsize = ftsz)
plt.xticks(range(0, 41, 10), ['0', '5M', '10M', '15M', '20M'], fontsize = 18)
plt.title('lambda=0.4, Lmt_ppd=0.30', fontsize = 18)

ax6 = plt.subplot(gs1[5])
ax6.set_xlim(0, 41)
ax6.plot(x, df['case12'], 'k', label='Case12 (act1)')
ax6.plot(x, df['case15'], 'k--',label='Case15 (act2)')
ax6.plot(x, df['case18'], 'k:', label='Case18 (act3)')
ax6.legend(fontsize = 14)
plt.xlabel('Training Step', fontsize = ftsz)
plt.xticks(range(0, 41, 10), ['0', '5M', '10M', '15M', '20M'], fontsize = 18)
plt.title('lambda=0.4, Lmt_ppd=1.0', fontsize = 18)

plt.setp(ax1.get_xticklabels(), visible=False)
plt.setp(ax2.get_xticklabels(), visible=False)
plt.setp(ax3.get_xticklabels(), visible=False)

ytk_ftsz = 14
plt.setp(ax1.get_yticklabels(), fontsize = ytk_ftsz)
plt.setp(ax2.get_yticklabels(), fontsize = ytk_ftsz)
plt.setp(ax3.get_yticklabels(), fontsize = ytk_ftsz)
plt.setp(ax4.get_yticklabels(), fontsize = ytk_ftsz)
plt.setp(ax5.get_yticklabels(), fontsize = ytk_ftsz)
plt.setp(ax6.get_yticklabels(), fontsize = ytk_ftsz)
```

```python
plt.show()
```
